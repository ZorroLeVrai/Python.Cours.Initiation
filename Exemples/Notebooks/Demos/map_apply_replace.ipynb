```python
import pandas as pd
from utilities import generer_dataframe
```

```python
df_participants = generer_dataframe(20)
df_participants
```

```python
df_participants["Nom_Prenom"] = df_participants.apply(lambda row: f"{row['Prenom']} {row['Nom']}", axis="columns")
df_participants
```

```python
df_participants.info()
```

```python
def generer_categorie(age):
    if age < 30:
        return "[18 - 29]"
    if age < 40:
        return "[30 - 39]"
    if age < 50:
        return "[40 - 49]"
    return "[50 - 65]"

df_participants["Catégorie"] = df_participants["Age"].apply(generer_categorie)
df_participants
```

```python
df_participants = df_participants.drop("Catégorie", axis="columns")
df_participants
```

```python
df_participants["Nouvelle Activite"] = df_participants["Activite"].map({"Tennis": "Padel", "Football": "Rugby"}, na_action="ignore")
df_participants
```

```python
df_participants["Categorie"] = df_participants["Age"].map(generer_categorie, na_action="ignore")
df_participants
```

```python
df_participants = df_participants.drop("Nouvelle Activite", axis="columns")
df_participants
```

```python
df_participants["Nouvelle Activite"] = df_participants["Activite"].replace(["Tennis", "Football"], ["Padel", "Rugby"])
df_participants
```
