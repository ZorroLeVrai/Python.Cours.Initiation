```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Donn√©es
x = np.linspace(0, 10, 101)
y = np.sin(x)

print("x=", x)
print("y=", y)
```

```python
# Tracer la courbe
plt.plot(x, y, label='sin(x)')
plt.title('Courbe sin(x)')
plt.xlabel('x')
plt.ylabel('sin(x)')

plt.legend()
plt.show()
```

```python
#donn√©es
x = np.arange(0, 10, 0.1)

print("x=", x)

# Plusieurs courbes sur le m√™me graphique
y1 = np.sin(x)
y2 = np.cos(x)

print(y2)
```

```python
plt.plot(x, y1, label='sin(x)', color='blue')
plt.plot(x, y2, label='cos(x)', color='red')
plt.title('Courbes sin(x) et cos(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True, which='both')

plt.axhline(y=0, color='k')
plt.axvline(x=0, color='k')

plt.legend()
plt.show()
```

```python
# Tracer des histogrammes avec Matplotlib
# Donn√©es
donnees = np.random.randn(1000)

# Tracer l'histogramme
plt.hist(donnees, bins=30, color='skyblue', edgecolor='black')
plt.title('Histogramme des donn√©es al√©atoires')
plt.xlabel('Valeurs')
plt.ylabel('Fr√©quence')
plt.show()
```

```python
# Donn√©es pour le Pie Chart
labels = ['Cat√©gorie 1', 'Cat√©gorie 2', 'Cat√©gorie 3']
sizes = [30, 45, 25]

# Tracer le Pie Chart
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
plt.title('Exemple de Pie Chart')
plt.show()
```
