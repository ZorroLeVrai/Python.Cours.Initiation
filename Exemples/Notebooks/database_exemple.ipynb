```python
import pandas as pd
from sqlalchemy import create_engine, text
```

```python
# Definition des paramètres de connexion à la base de données
db_username = 'postgres'
db_password = 'pass123'
db_host = '127.0.0.1'
db_port = '5432'
db_name = 'postgres'
```

```python
# Création du moteur SQLAlchemy
engine = create_engine(f'postgresql+psycopg2://{db_username}:{db_password}@{db_host}:{db_port}/{db_name}')
```

```python
# Exécution d'une requête simple
sql_query = "SELECT * FROM ville"
df_villes = pd.read_sql(sql_query, engine)
df_villes
```

```python
# Lister toutes les tables de PostgreSQL
sql_query = "SELECT tablename FROM pg_catalog.pg_tables;"
df_villes = pd.read_sql(sql_query, engine)
df_villes["tablename"].to_list()
```

```python
# Exécution d'une requête paramètrée
departement_a_rechercher = "60"

sql_query = text("SELECT * FROM public.ville WHERE ville_departement = :departement_code")
params = {"departement_code": departement_a_rechercher}
df_villes_departement = pd.read_sql_query(sql_query, engine, params=params)
df_villes_departement
```

```python
# Affichage des 100 première lignes
df_villes_100 = df_villes.head(100)
df_villes_100
```

```python
from datetime import datetime

def obtenir_date_heure() -> str:
    date_heure_format = "%d/%m/%Y %H:%M:%S"
    now = datetime.now()
    return now.strftime(date_heure_format)
```

```python
# Sauvegarde dans Excel les 100 premières lignes
fichier_de_sauvegarde = "output_villes.xlsx"
df_villes_100.to_excel(fichier_de_sauvegarde, index=False)
print("Les données ont été sauvegardées dans le fichier", fichier_de_sauvegarde)
print("Date et heure de sauvegarde", obtenir_date_heure())
```

```python
# Sauvegarde dans une table de base de données
nom_table = "ville_output"
df_villes_departement.to_sql(nom_table, engine, if_exists='replace', index=False)
print("Les données ont été sauvegardées dans la table", nom_table)
print("Date et heure de sauvegarde", obtenir_date_heure())
```
