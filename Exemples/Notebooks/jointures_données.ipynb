```python
import pandas as pd
```

```python
# Cr√©ation des dataframes
df1 = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'Name': ['Alice', 'Bob', 'Charlie', 'David']
})

df2 = pd.DataFrame({
    'ID': [3, 4, 5, 6],
    'Score': [85, 90, 95, 100]
})
```

```python
print("Dataframe 1")
display(df1)

print("Dataframe 2")
display(df2)
```

```python
# Inner join
df_result = pd.merge(df1, df2, on='ID')
display(df_result)
```

```python
# Left join
left_join = pd.merge(df1, df2, on='ID', how='left')
left_join
```

```python
# Right join
right_join = pd.merge(df1, df2, on='ID', how='right').fillna(0)
right_join
```

```python
# Outer join
outer_join = pd.merge(df1, df2, on='ID', how='outer').fillna(0)
outer_join
```

```python
# Jointure sur plusieurs colonnes
df3 = pd.DataFrame({
    'ID': [1, 2, 3, 4],
    'Subject': ['Math', 'Science', 'English', 'History'],
    'Name': ['Alice', 'Bob', 'Charlie', 'David']
})

df4 = pd.DataFrame({
    'ID': [3, 4, 1, 2],
    'Subject': ['English', 'History', 'Math', 'Science'],
    'Score': [85, 90, 95, 100]
})

# Jointure utilisant les colonnes 'ID' et 'Subject'
multi_join = pd.merge(df3, df4, on=['ID', 'Subject'])
multi_join
```
